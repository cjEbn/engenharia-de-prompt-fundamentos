```python
# 1. Conectar ao Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 2. Bibliotecas
import os
import shutil

# 3. Ver pastas disponíveis no Drive
print("📁 Pastas no seu Google Drive:")
print(os.listdir("/content/drive/MyDrive"))

# 4. ESCOLHE AQUI UMA PASTA QUE APARECEU ACIMA
pasta_origem = "/content/drive/MyDrive/Colab Notebooks"  # MUDA SE PRECISAR

# 5. Verificar se a pasta existe
if not os.path.exists(pasta_origem):
    print("❌ Pasta não encontrada! Verifique o nome.")
else:
    print("✅ Pasta encontrada! Iniciando organização...\n")

    tipos_arquivos = {
        "PDF": [".pdf"],
        "Imagens": [".jpg", ".jpeg", ".png", ".gif"],
        "Documentos": [".docx", ".txt", ".pptx"],
        "Planilhas": [".xlsx", ".csv"],
        "Outros": []
    }

    def organizar_arquivos(pasta):
        total_movidos = 0

        for arquivo in os.listdir(pasta):
            caminho_arquivo = os.path.join(pasta, arquivo)

            if os.path.isdir(caminho_arquivo):
                continue

            nome, extensao = os.path.splitext(arquivo)
            extensao = extensao.lower()

            movido = False

            for pasta_destino, extensoes in tipos_arquivos.items():
                if extensao in extensoes:
                    destino = os.path.join(pasta, pasta_destino)
                    os.makedirs(destino, exist_ok=True)

                    try:
                        shutil.move(caminho_arquivo, os.path.join(destino, arquivo))
                        print(f"Movido: {arquivo} → {pasta_destino}")
                        total_movidos += 1
                    except PermissionError:
                        print(f"Erro de permissão: {arquivo}")

                    movido = True
                    break

            if not movido:
                destino = os.path.join(pasta, "Outros")
                os.makedirs(destino, exist_ok=True)

                try:
                    shutil.move(caminho_arquivo, os.path.join(destino, arquivo))
                    print(f"Movido: {arquivo} → Outros")
                    total_movidos += 1
                except PermissionError:
                    print(f"Erro de permissão: {arquivo}")

        print(f"\n✅ Organização concluída! {total_movidos} arquivos movidos.")

    # 6. Executar
    organizar_arquivos(pasta_origem)
```
