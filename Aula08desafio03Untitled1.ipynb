```python
import sys

# Função para verificar condição crítica com limite configurável
def verificar_temperatura(temp, limite_critico=50):
    """
    Verifica se a temperatura está acima de um limite crítico.

    Args:
        temp (float): A temperatura atual.
        limite_critico (int, optional): O limite de temperatura. Padrão é 50.

    Returns:
        bool: True se a temperatura for crítica, False caso contrário.
    """
    return temp > limite_critico

# Função para enviar alerta
def enviar_alerta(temp):
    """
    Exibe uma mensagem de alerta para temperatura crítica.

    Args:
        temp (float): A temperatura crítica detectada.
    """
    print(f"⚠️ ALERTA! Temperatura crítica detectada: {temp}°C")

# Função principal do sistema de monitoramento
def sistema_monitoramento(limite_critico=50):
    """
    Executa o sistema de monitoramento de temperatura.

    Args:
        limite_critico (int, optional): O limite de temperatura. Padrão é 50.
    """
    while True:
        try:
            temperatura_str = input("Digite a temperatura atual (ou 'sair' para encerrar): ")
            if temperatura_str.lower() == 'sair':
                print("Encerrando o sistema de monitoramento.")
                break

            temperatura = float(temperatura_str)

            if verificar_temperatura(temperatura, limite_critico):
                enviar_alerta(temperatura)
            else:
                print("✅ Temperatura normal.")
            break # Sai do loop se a entrada for válida
        except ValueError:
            print("❌ Erro: Por favor, digite um número válido para a temperatura.")
        except Exception as e:
            print(f"❌ Ocorreu um erro inesperado: {e}")

# Ponto de entrada do programa
if __name__ == "__main__":
    # Você pode configurar o limite crítico aqui, por exemplo, para 60 graus:
    # sistema_monitoramento(limite_critico=60)
    sistema_monitoramento()
```
