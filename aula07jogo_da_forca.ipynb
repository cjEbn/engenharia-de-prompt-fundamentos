```python
import random

def choose_word(word_list):
    """Escolhe uma palavra aleatória da lista."""
    return random.choice(word_list)

def display_game_state(word_display, attempts_left, guessed_wrong_letters):
    """Exibe o estado atual do jogo: palavra, tentativas restantes e letras erradas."""
    print(f"\nPalavra: {' '.join(word_display)}")
    print(f"Tentativas restantes: {attempts_left}")
    print(f"Letras erradas: {', '.join(sorted(guessed_wrong_letters))}")

def get_player_guess(already_guessed_letters):
    """Solicita e valida a entrada da letra do jogador, evitando repetições."""
    while True:
        guess = input("Digite uma letra: ").lower()
        if not guess.isalpha():
            print("Por favor, digite apenas letras.")
        elif len(guess) != 1:
            print("Por favor, digite apenas uma letra por vez.")
        elif guess in already_guessed_letters:
            print(f"Você já tentou a letra '{guess}'. Tente outra.")
        else:
            return guess

def update_word_display(word, word_display, guess):
    """Atualiza a exibição da palavra com letras corretas."""
    for i in range(len(word)):
        if word[i] == guess:
            word_display[i] = guess

def play_hangman():
    """Função principal que executa o jogo da forca."""
    word_list = ["python", "computador", "engenharia", "software", "programacao", "algoritmo", "desenvolvimento", "inteligencia"]
    word = choose_word(word_list)
    word_display = ["_"] * len(word)
    attempts_left = 6
    guessed_letters = set() # Contém todas as letras já tentadas (certas e erradas)
    guessed_wrong_letters = [] # Contém apenas as letras erradas para exibir

    print("🎮 Jogo da Forca")

    while attempts_left > 0 and "_" in word_display:
        display_game_state(word_display, attempts_left, guessed_wrong_letters)

        current_guess = get_player_guess(guessed_letters)
        guessed_letters.add(current_guess)

        if current_guess in word:
            update_word_display(word, word_display, current_guess)
            print(f"Boa! A letra '{current_guess}' está na palavra.")
        else:
            attempts_left -= 1
            guessed_wrong_letters.append(current_guess)
            print(f"Que pena! A letra '{current_guess}' não está na palavra. Você tem {attempts_left} tentativas restantes.")

    # Fim do jogo
    print(f"\nPalavra final: {' '.join(word_display)}")
    if "_" not in word_display:
        print(f"🏆 Parabéns! Você venceu! A palavra era: {word.upper()}")
    else:
        print(f"💀 Que pena! Você perdeu! A palavra correta era: {word.upper()}")

# Inicia o jogo
if __name__ == "__main__":
    play_hangman()
```
