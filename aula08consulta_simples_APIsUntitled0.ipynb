```python
import requests
import json # Import the json module to catch JSONDecodeError

def consultar_cep(cep):
    try:
        # URL CORRIGIDA: 'viacep' ao invés de 'vacep'
        url = f"https://viacep.com.br/ws/{cep}/json/"

        resposta = requests.get(url)

        # Força erro se a requisição falhar (status code diferente de 2xx)
        resposta.raise_for_status()

        dados = resposta.json()

        if "erro" in dados and dados["erro"] == True:
            print("❌ CEP não encontrado.")
        else:
            print("📍 Endereço encontrado:")
            print(f"Rua: {dados.get('logradouro', 'N/A')}")
            print(f"Bairro: {dados.get('bairro', 'N/A')}")
            print(f"Cidade: {dados.get('localidade', 'N/A')}")
            print(f"Estado: {dados.get('uf', 'N/A')}")

    except requests.exceptions.HTTPError as e:
        print(f"⚠️ Erro HTTP: {e.response.status_code} - {e.response.reason}")
    except requests.exceptions.ConnectionError as e:
        print("⚠️ Erro de conexão. Verifique sua internet ou a disponibilidade do serviço.")
    except requests.exceptions.Timeout as e:
        print("⚠️ A requisição excedeu o tempo limite.")
    except json.JSONDecodeError as e:
        print(f"⚠️ Erro ao decodificar a resposta JSON: A resposta da API não é um JSON válido. Detalhes: {e}")
    except requests.exceptions.RequestException as e:
        print("⚠️ Erro inesperado na requisição:", e)
    except Exception as e:
        print("⚠️ Ocorreu um erro desconhecido:", e)

cep = input("Digite o CEP: ")
consultar_cep(cep)
```
